# Import Financial Data in to Python

```python
import yfinance as yf
stock_data = yf.download("AAPL")
display(stock_data)
```

```python
stock_data = yf.download("AAPL", start="2022-01-01", end="2024-04-01")
display(stock_data)
```

```python
yf.download("AAPL", period="1y")
```

```python
yf.download("AAPL", period="3d", interval="1m")
```

```python
tickers = ["aapl", "msft", "goog"]
yf.download(tickers, period="1Y")
```

```python
apple = yf.Ticker("AAPL")
display(apple.info)
```

```python
import pandas as pd
apple_df = pd.json_normalize(apple.info)
display(apple_df)
```

```python
display(apple.info['priceToBook'])
```

```python
display(apple_df['priceToBook'])
```

```python
display(apple.history(period="1Y"))
```

```python
display(apple.dividends)
```
